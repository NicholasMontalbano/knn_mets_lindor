```python
import pandas as pd
import numpy as np
import psycopg2
from psycopg2 import OperationalError
import matplotlib.pyplot as plt
```

```python
similiarity = pd.read_csv('../data/tables/similiarity_final_age_mlb.csv')
lind = similiarity.loc[similiarity['og_playerid'] == 'lindofr01']
playerid = tuple(lind['comp_playerid'].to_list() + lind['og_playerid'].drop_duplicates().to_list())
playerid 
```

```python
def create_connection(db_name, db_user, db_password, db_host, db_port):
    connection = None
    try:
        connection = psycopg2.connect(
            database=db_name,
            user=db_user,
            password=db_password,
            host=db_host,
            port=db_port,
        )
        print("Connection to PostgreSQL DB successful")
    except OperationalError as e:
        print(f"The error '{e}' occurred")
    return connection

connection = create_connection(
    "nicholasmontalbano", "postgres", "", "127.0.0.1", "5432"
)

curs = connection.cursor()
```

```python
curs.execute("ROLLBACK")
```

```python
connection.commit()
```

```python
sql_command = """
SELECT * FROM

(SELECT * 
FROM batting_mlb
WHERE playerid IN {}) a

LEFT JOIN (
SELECT playerID, birthYear, birthMonth, birthDay, nameFirst, nameLast
FROM people_mlb) b

USING(playerID)
""".format(playerid)

comp = pd.read_sql(sql_command, connection)
```

```python
comp = comp.fillna(0)
comp["avg"] = comp["h"] / comp["ab"]
comp["slg"] = ((comp["h"] - comp["double"] - comp["triple"] - comp["hr"]) + (comp["double"]*2) + (comp["triple"]*3) + (comp["hr"]*4)) / comp["ab"]
comp['obp'] = (comp['h'] + comp['bb'] + comp['ibb'] + comp['hbp']) / (comp['ab'] + comp['bb'] + comp['hbp'] + comp['sf'])
comp['ops'] = comp['obp'] + comp['slg']
comp['yearid'] = pd.to_numeric(comp['yearid'])
comp['age'] = comp['yearid'] - comp['birthyear']
comp
```

```python
fig = plt.figure()

fig, ax = plt.subplots(figsize=(10,8))
comp.loc[(comp['age']<=28) & (comp['playerid'] == 'lindofr01')].sort_values('age').plot(ax=ax, x='age', y='ops', c='r')
comp.loc[(comp['age']<=28) & (comp['playerid'] == 'baezja01')].sort_values('age').plot(ax=ax, x='age', y='ops', c='b')
comp.loc[(comp['age']<=28) & (comp['playerid'] == 'bryankr01')].sort_values('age').plot(ax=ax, x='age', y='ops', c='y')
comp.loc[(comp['age']<=28) & (comp['playerid'] == 'bogaexa01')].sort_values('age').plot(ax=ax, x='age', y='ops', c='g')

plt.axvline(x=27, linestyle='--')
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ylim(0,1.0)
```
